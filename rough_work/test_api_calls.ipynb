```python
import requests
```

```python
url = "https://statsapi.mlb.com/api/v1/schedule"
params = {
    "sportId": 1,
    "season": 2024,
    "gameType": "R"
}
# R is regular season
```

```python
response = requests.get(url, params=params)
data = response.json()
print(data.keys())
```

```python
dates = data.get("dates", [])
print(type(dates), len(dates))

# inspect first element
if dates:
    print(dates[0].keys())

    games = dates[0].get("games", [])
    if games:
        print(games[0].keys())
```
