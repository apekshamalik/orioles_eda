```python
import requests
```

```python
url = "https://statsapi.mlb.com/api/v1/schedule"
params = {
    "sportId": 1,
    "season": 2024,
    "gameType": "R"
}
# R is regular season
```

```python
response = requests.get(url, params=params)
data = response.json()
print(data.keys())
```

```python
dates = data.get("dates", [])
print(type(dates), len(dates))

# inspect first element
if dates:
    print(dates[0].keys())

    games = dates[0].get("games", [])
    if games:
        print(games[0].keys())
```

```python
import json
r = requests.get('https://statsapi.mlb.com/api/v1/teams', params={
    'sportId': 1,
    'season': 2024,
    'hydrate': 'league,division'
})

print(f"Status: {r.status_code}")
data = r.json()
print(f"Keys: {data.keys()}")
teams = data.get("teams", [])
print(f"Teams count: {len(teams)}")
print(json.dumps(teams[0], indent=2))
```

```python
r = requests.get('https://statsapi.mlb.com/api/v1/schedule', params={
    'sportId': 1,
    'season': 2024,
    'gameType': 'R'
})

data = r.json()
dates = data.get("dates", [])
first_game = dates[0]["games"][0]
print(f"Total date buckets: {len(dates)}")
print(json.dumps(first_game, indent=2))
```

```python
import requests, json

game_pk = 745444 

r = requests.get(f'https://statsapi.mlb.com/api/v1.1/game/{game_pk}/feed/live')
data = r.json()

#at-bat
plays = data["liveData"]["plays"]["allPlays"]
print(f"Total plays in game: {len(plays)}")

#one-play
play = plays[0]
print(json.dumps({
    "atBatIndex": play.get("atBatIndex"),
    "result": play.get("result"),
    "about": play.get("about"),
    "matchup": play.get("matchup")
}, indent=2))

#pitch event
for event in play.get("playEvents", []):
    if "pitchData" in event:
        print(json.dumps(event, indent=2))
        break
```
